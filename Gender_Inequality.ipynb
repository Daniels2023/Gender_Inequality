```python
import pandas as pd
```

```python
df_Gender = pd.read_csv('./Resources/Gender.csv', encoding='cp1252')
df_Gender
```

```python
df_Gender = df_Gender.drop(columns={'GII VALUE' ,'GII RANK','Maternal_mortality','Adolescent_birth_rate'
})
```

```python
df_Gender = df_Gender.rename(columns={'HUMAN DEVELOPMENT':'Human_Development',
                          'Seats_parliamentt(% held by women)':'Seats_in_Parliament',
                          'F_secondary_educ':'Female_Education',
                          'M_secondary_educ':'Male_Education',
                          'F_Labour_force':'Female_Workforce',
                          'M_Labour_force':'Male_Workforce',
                          'HDI rank':'HDI_rank'

})
```

```python
df_Gender = df_Gender.dropna()
```

```python
df_Gender.reset_index(inplace=True, drop=True)
df_Gender.index += 1
```

```python
df_Gender = df_Gender.drop(columns = ['HDI_rank'])
```

```python
df_Gender = df_Gender.reset_index()
```

```python
df_Gender = df_Gender.rename(columns={'index': 'HDI_rank'})
```

```python
df_Gender = df_Gender.set_index('HDI_rank')
```

```python
df_Gender.to_csv('./Resources/GenderCleaned.csv')
```
